# Perceptron trick 

```python
from sklearn.datasets import make_classification
import numpy as np
X,y=make_classification(n_samples=100,n_features=2,n_informative=1,n_redundant=0,
                       n_classes=2,n_clusters_per_class=1,random_state=41,hypercube=False,class_sep=30)

import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
```

```python
def perceptron(X,y):
    X=np.insert(X,0,1,axis=1)
    weights=np.ones(X.shape[1])
    lr=0.1
    for i in range(1000):
        j=np.random.randint(0,100)
        y_hat=step(np.dot(X[j],weights))
        weights=weights + lr * (y[j]-y_hat)*X[j]
    return weights[0],weights[1:]
```

```python
def step(z):
    return 1 if z>=0 else 0
```

```python
intercept_ ,coef_ = perceptron(X,y)
```

```python
print(intercept_ , coef_ )
```

```python
m=-(coef_[0]/coef_[1])
b=-(intercept_/coef_[1])
x_input=np.linspace(-3,3,100)
y_input=m*x_input+b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```

# Voting-ensemle classifier



```python
'''A voting ensemble is a type of ensemble learning method where multiple individual models are trained and their
predictions are combined to make a final prediction '''

'''
Hard Voting:

In hard voting, each model in the ensemble makes a prediction, and the final prediction is determined by a majority vote.
The class that receives the majority of votes is the predicted class.
Hard voting is typically used for classification problems.

Soft Voting:

In soft voting, each model in the ensemble provides a probability distribution over the classes. 
The final prediction is then based on the average of these probability distributions.
Soft voting can be used for both classification and regression problems.

'''

'''
d->m1  <--x(query)------>trained->>>>value =y1
d->m2  <--x(query)------>trained->>>>value =y2
d->m3  <--x(query)------>trained->>>>value =y3   
 took majority of these all and reach to the final answer
 
'''

import pandas as pd
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt

encoder=LabelEncoder()

iris_data=pd.read_csv("iris.csv")
iris_data=iris_data.drop(columns=['Id'])
# iris_data.head()
iris_data['Species']=encoder.fit_transform(iris_data['Species'])
iris_data.head()

sns.pairplot(iris_data,hue='Species')

```

```python
print(iris_data['Species'].value_counts())
print(len(iris_data['Species']))
# plt.bar(iris_data['Species'].value_counts().keys(),iris_data['Species'].value_counts())
iris_data_final = iris_data[iris_data['Species'] !=0 ][['SepalLengthCm','SepalWidthCm','Species']]
iris_data_final.head()
plt.bar(iris_data_final['Species'].value_counts().keys(),iris_data_final['Species'].value_counts())
```

```python
X= iris_data_final.iloc[:,0:2]
y= iris_data_final.iloc[:,-1]
```

```python
'''my models will be 1. KNN , 2. random forest, 3. logistic regression'''
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

```python
clf1=KNeighborsClassifier()
clf2=LogisticRegression()
clf3=RandomForestClassifier()
```

```python
estimators=[('knn',clf1),('lr',clf2),('rf',clf3)]

for estimator in estimators:
    x=cross_val_score(estimator[1],X,y,cv=10,scoring='accuracy')
#     print(x)
    print(estimator[0],np.round(np.mean(x),2))
```

```python
from sklearn.ensemble import VotingClassifier
vc= VotingClassifier(estimators=estimators,voting='hard')
x=cross_val_score(vc,X,y,cv=10,scoring='accuracy')
print(np.round(np.mean(x),2))
```

```python
vc1=VotingClassifier(estimators=estimators,voting='soft')
x=cross_val_score(vc1,X,y,cv=10,scoring='accuracy')
print(np.round(np.mean(x),2))
```

# voting ensemble regression

```python
'''
d->m1  <--x(query)------>trained->>>>value =y1
d->m2  <--x(query)------>trained->>>>value =y2
d->m3  <--x(query)------>trained->>>>value =y3   
 took mean of these all and reach to the final answer
 
'''
```

```python
from sklearn.datasets import load_diabetes  #sklean linear data
from sklearn.linear_model import LinearRegression 
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
import numpy as np
X,y=load_diabetes(return_X_y=True)

lr=LinearRegression()
dt=DecisionTreeRegressor()
svr=SVR()

estimators=[('lr',lr),('dt',dt),('svr',svr)]

for estimator in estimators:
    scores=cross_val_score(estimator[1],X,y,cv=10,scoring='r2')
    print(estimator[0],np.round(np.mean(scores),2))
```

```python
from sklearn.ensemble import VotingRegressor
vr=VotingRegressor(estimators)
score=cross_val_score(vr,X,y,cv=10,scoring='r2')
print("Voting regressor",np.round(np.mean(score),2))
```
